### Cleaning demographics columns
**My Key**  
Grade in school  
>numerical 6 through 12  

Sex:  
>    0 = female  
>    1 = male   

Race/ethnicity columns - dummy variables:  
>    0 = no  
>    1 = yes  
    *note that since students were prompted to select more than 1 option (when applicable), I am not dropping a column (as you normally would when creating dummy variables)   

```python
# selecting demographic columns of interest

ecig_demographic = nyts_full.loc[:, ['QN3', 'QN2']]
ecig_demographic.head()
```

```python
# converting QN3 to 'grade'
ecig_demographic['grade'] = ecig_demographic['QN3'].apply(lambda x: np.nan if x == 8 else x+5)

# converting QN2 to "sex" composed of 0 (female) & 1s (male)
ecig_demographic['sex'] = ecig_demographic['QN2'].apply(lambda x: 0 if x == 2 else x)

# removing unneeded columns
ecig_demographic = ecig_demographic.drop(['QN3', 'QN2'], axis=1)
ecig_demographic.head()
```

```python
# selecting race/ethnicity columns

ecig_race = nyts_full.loc[:,['QN5A', 'QN5B', 'QN5C', 'QN5D', 'QN5E', \
                                      'QN4A','QN4B', 'QN4C', 'QN4D', 'QN4E']
                                   ]
ecig_race.head()
```

```python
# function for combining dummy variable columns

def combine_hispanic_columns(col0, col1, col2, col3, col4):
    '''Combines multiple dummy variable columns representing hispanic sub-types into a 
    single column'''
    
    if col1 == 1 or col2 == 1 or col3 == 1 or col4 == 1:
        return 1
    elif col0 == 1:
        return 0
    else:
        return np.nan
```

```python
#  combining hispanic sub-types (QN4B, C, D, E) and not-hispanic (QN4A) into 1 column
ecig_race['hispanic'] = ecig_race.apply(lambda x: combine_hispanic_columns(x['QN4A'], x['QN4B'], x['QN4C'], \
                                            x['QN4D'], x['QN4E']), axis=1)

# removing unneeded hispanic columns
ecig_race = ecig_race.drop(['QN4A','QN4B', 'QN4C', 'QN4D', 'QN4E'], axis=1)
ecig_race.head()
```

```python
# changing race/ethnicity column names 

ecig_race.columns = ['native', 'asian', 'black', 'pac_island', 'white', 'hispanic']
ecig_race.head()
```

```python
def replace_race_nulls(current_col, col1, col2, col3, col4, col5, col6):
    '''Replaces nulls in race/ethnicity columns - unless race/ethnicity is not entered, 
    in which case the null stays'''
    
    if current_col == 1:
        return 1
    elif col1 == 1 or col2 == 1 or col3 == 1 or col4 ==  1 or col5 == 1 or col6 == 1:
        return 0
    else:
        return np.nan
```

```python
# replace nulls in each race/ethicity column

for col in ecig_race.columns:
    ecig_race[col] = ecig_race.apply(lambda x: replace_race_nulls(x[col], 
            x['native'], x['asian'], x['black'], 
            x['pac_island'], x['white'], x['hispanic']), axis=1)

    ecig_race.head()
```

```python
ecig_demographic = pd.concat([ecig_demographic, ecig_race], axis=1)
ecig_demographic.head()
```

### Cleaning tobacco use columns
**My Key**  
Ever used e-cig:    
Ever smoked a cigarette:   
Ever smoked cigars or cigarillos:   
Ever used other tobacco product:   
Ever vaped marijuana:   
>    0 = no  
>    1 = yes 

```python
# selecting columns for use of e-cigs, cigarettes, cigars,  marijuana vape
# e-cig column will be the target

tobacco_mj = nyts_full.loc[:, ['QN6', 'QN22', 'QN35', 'QN21']]
tobacco_mj.head()
```

```python
# converting column data to 1s (yes) and 0s (no)
for col in tobacco_mj.columns[:3]:
    tobacco_mj[col] = tobacco_mj[col].apply(lambda x: 0 if x==2 else x)

# for mj column, combining "don't know" into "no"
tobacco_mj['QN21'] = tobacco_mj['QN21'].apply(lambda x: 0 if (x == 2) | (x == 3) else x)

tobacco_mj.columns = ['e_cig', 'smoke_cig', 'cigar', 'vape_mj']
tobacco_mj.head()
```

```python
# selecting columns relating to other types of tobacco use (hookah, chewing tobacco, etc)

other_tobacco = nyts_full.loc[:, ['QN44', 'QN48', 'QN56', 'QN53A', 'QN53B', 'QN53C', 'QN53D', 'QN53E']]
other_tobacco.head()
```

```python
# converting to 1s and 0s

for col in other_tobacco.columns[:2]:
    other_tobacco[col] = other_tobacco[col].apply(lambda x: 0 if x==2 else x)

other_tobacco['QN56'] = other_tobacco['QN56'].apply(lambda x: 0 if (x==2) | (x==3) else x)
other_tobacco.head()
```

```python
# combining 'other_tobacco' into a single column - a 1 in any column will lead to a value of 1

other_tob = other_tobacco.max(axis=1)
other_tob.head()
```

```python
# adding other_tobacco to tobacco/mj df

tobacco_mj.insert(3, 'other_tob', other_tob)
tobacco_mj.head()
```

### Cleaning marketing columns
**My Key**   
- Internet ads (when using the internet how often do you see e-cig ads?)
- TV/streaming ads (when watching TV or streaming services how often do you see e-cig ads?)  
- Social media ads (when using social media how often do you see e-cig-related posts?)    
>1 = Never/I do not use the (platform)    
>2 = Rarely  
>3 = Sometimes 
>4 = Most of the time  
>5 = Always  

- Internet use (do you use the internet?)  
- TV/streaming use (do you use TV/streaming services?)  
- Social media use (do you use social media?)  
>0 = no    
>1 = yes  

```python
# selecting & renaming relevant marketing columns
# these columns have (only) 300-400 nulls but 3 of them are in the first 5 rows!

marketing = nyts_full.loc[:, ['QN102', 'QN105', 'QN106']]
marketing.columns = ['internet_ads', 'TV/stream_ads', 'social_ads']
marketing.head()
```

```python
def get_platform_use(col):
    '''Creates platform use (yes/no) columns from marketing data'''
    if col == 1:
        return 0
    elif col > 1:
        return 1
    else:
        np.nan
```

```python
# creating new set of columns for platform use (yes/no)

marketing['internet_use'] = marketing['internet_ads'].apply(get_platform_use)
marketing['TV/stream_use'] = marketing['TV/stream_ads'].apply(get_platform_use)
marketing['social_use'] = marketing['social_ads'].apply(get_platform_use)
```

```python
marketing.head()
```

```python
# adjusting numbering for ad columns, combining 'doesn't use platform' with 'never sees ads'

for col in marketing.columns[:3]:
    marketing[col] = marketing[col].apply(lambda x: 1 if x == 1 else x-1)
```

```python
marketing.head()
```

### Cleaning attitude columns
**My Key**   
QN89 - e-cig addictiveness (do you believe e-cigs are...?)  
>-1 = less addictive  
>0 = equally addictive/never heard of e-cigs/don't know enough  
>+1 = more addictive     

QN88 - harm 2/2 e-cigs (how much do people harm themselves when using e-cigs?)      
QN95 - harm 2/2 e-cig vapor (how harmful is breathing the vapor from someone else's e-cig?)      
>1 = no harm  
>2 = little harm  
>3 = some harm  
>4 = a lot of harm  

```python
# selecting & renaming attitude columns

attitude = nyts_full.loc[:, ['QN89', 'QN88', 'QN95']]
attitude.columns = ['addictive', 'harm_self', 'harm_others']
attitude.head()
```

```python
def reorganize_addictive_col(col):
    '''Re-organizes addictive column by making 'less' = -1 and 'more' = 1; also
    combines 'equal', 'never heard of e-cigs' and 'don't know enough' into 0
    '''
    
    if col == 1:
        return -1
    elif col == 2 or col == 4 or col == 5:
        return 0
    elif col == 3:
        return 1
    elif col == np.nan:
        return np.nan
```

```python
attitude['addictive'] = attitude['addictive'].apply(reorganize_addictive_col)
attitude.head()
```

### Cleaning other columns
**My Key**      
QN82 - "The Real Cost" (in the past 12 months, have you seen or heard the "The Real Cost' in any form (TV, internet, social, radio, etc)) - "The Real Cost" is FDA's teen tobacco prevention program
>0 = no/not sure  
>1 = yes

QN76 - healthcare (in the past 12 months, has a healtchare professional [doctor, dentist, nurse, etc] asked you if you use e-cigarettes?) split into 2 columns:  
- health prof asked (did a healthcare professional (doctor, dentist, nurse, etc) ask you if you use e-cigs within the last 12 months?  
- health prof seen (Did you see a healthcare professional within the past 12 months?):  
>- 0 = no  
>- 1 = yes    

QN114 - tobacco at home (does anyone who lives with you use tobacco products?)  
>0 = no  
>1 = yes  


```python
# selecting and renaming two "other" columns (real cost and healthcare provider)

other = nyts_full.loc[:, ['QN82', 'QN76']]
other.columns = ['real_cost', 'healthcare']
other.head()
```

```python
# converting 'real_cost' into 0s (no/not sure) and 1s(yes)

other['real_cost'] = other['real_cost'].apply(lambda x: 0 if (x==2) | (x==3) else x)
other.head()
```

```python
def get_health_seen(col):
    '''Creates a new column for "seen by health professional within 12 months" (yes/no)
    '''
    
    if col==1:
        return 0
    elif col==2 or col == 3:
        return 1
    elif col==np.nan:
        return np.nan
```

```python
other['health_seen'] = other['healthcare'].apply(get_health_seen)
other.head()
```

```python
def get_health_asked(col):
    '''Creates a new column for "asked about e-cigs by health professional within 
    12 months" (yes/no)
    '''
  
    if col==1 or col==3:
        return 0
    elif col==2:
        return 1
    elif col==np.nan:
        return np.nan
```

```python
other['health_asked'] = other['healthcare'].apply(get_health_asked)
other.drop(columns='healthcare', inplace=True)
other.head()
```

```python
# selecting tabacco at home columns - this was all one question with many answer choices
# only 484 participates did not answer the question, but 4 of them are in the first 5 rows!

tobacco_home = nyts_full.loc[:, ['QN114A', 'QN114B', 'QN114C', 'QN114D',
                  'QN114E', 'QN114F', 'QN114G', 'QN114H',
                 'QN114I', 'QN114J', 'QN114K']]
tobacco_home.head()
```

```python
# combining tobacco at home columns - so any type of tobacco use will be a 1 in 'yes' column

tobacco_home['yes'] = tobacco_home.iloc[:, :-1].max(axis=1)
tobacco_home.head()
```

```python
# QN114K represents a 'no tabacco at home' answer

tobacco_home = tobacco_home.rename({'QN114K': 'no'}, axis=1)
tobacco_home.head()
```

```python
def combine_home_tobacco(no, yes):
    '''Combines yes and no columns for "tobacco user at home" into a single column
    '''
    
    if yes == 1:
        return 1
    elif no == 1:
        return 0
    else:
        return np.nan
```

```python
tob_home = tobacco_home.apply(lambda x: combine_home_tobacco(x['no'], x['yes']), axis=1)
tob_home.head()
```

```python
# adding tob_home to "other" df

other['tob_home'] = tob_home
other.head()
```

## Combining columns 

```python
# combining subset dfs into a single df

ecig_df = pd.concat([tobacco_mj, ecig_demographic, marketing, attitude, other], axis=1)
ecig_df.head()
```

## Cleaning nulls 

```python
# in general, nulls increase towards the end of the survey - students probably gave up or ran out of time

ecig_df.isnull().sum()
```

> - Potentially concerned that dropping nulls would disproportionately drop students who use e-cigs - since those who answered 'yes' to e_cig use (or other tobacco use) were given additional questions   
> - But dropping all nulls didn't make a huge difference in the % of students who use e_cigs (25.6% to 24.5%)

```python
# prior to dropping nulls, 25.6% of students used e-cigs (our classification target)

ecig_df['e_cig'].value_counts()
```

```python
# if we were to drop all nulls, 24.5% of students use e-cigs

ecig_df.dropna()['e_cig'].value_counts()
```

```python
# dropping all nulls (~1400 out of 14506 data points) 

ecig_df = ecig_df.dropna().reset_index(drop=True)
ecig_df.head()
```
